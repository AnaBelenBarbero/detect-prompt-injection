```python
import pandas as pd
from dotenv import load_dotenv
from datasets import load_dataset

import os 


load_dotenv()

benign_code_en_prompts = load_dataset(
    "nomic-ai/gpt4all_prompt_generations", 
    split='train[:200]',
    #download_mode='force_redownload'
)

benign_en_prompts = pd.read_csv("hf://datasets/fka/awesome-chatgpt-prompts/prompts.csv")

benign_en_prompts_2 = pd.read_csv("hf://datasets/MohamedRashad/ChatGPT-prompts/train.csv")

benign_en_prompts_3 = pd.read_parquet("../data/raw/Prompt Engineering.parquet")
```

```python
benign_code_en_prompts = [i.lstrip("<p>").rstrip("</p>") for i in benign_code_en_prompts["prompt"]]
benign_en_prompts["prompt"].tolist()
benign_en_prompts_2["human_prompt"].tolist()
benign_en_prompts_3["Prompt"].tolist()
print("OK")
```

```python
df_concat = pd.DataFrame({
    "text": benign_code_en_prompts + benign_en_prompts["prompt"].tolist() + benign_en_prompts_2["human_prompt"].tolist() + benign_en_prompts_3["Prompt"].tolist(),
    "type": 0
})
print(df_concat.shape)
df_concat.head()
```

## Export English: *benign_2k_en_prompts.csv*


```python
df_concat.to_csv('../data/benign_2k_en_prompts.csv', index=False)
```

## Translate to Spanish: *benign_2k_es_prompts.csv*

```python
from googletrans import Translator
traslator = Translator()

translations = []
for i in df_concat["text"]:
    trans = await traslator.translate(i, src='en', dest='es')
    translations.append(trans.text)

translations
```

```python
len(translations)
```

```python
df_translations = pd.DataFrame({
    "texto": translations,
    "tipo": 0,
})

df_translations
```

```python
df_translations.to_csv('../data/benign_2k_es_prompts.csv', index=False)
```
